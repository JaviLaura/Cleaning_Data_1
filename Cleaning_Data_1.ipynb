```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

choleraDF = pd.read_csv("https://static.junilearning.com/python_data_science/cholera.csv")
display(choleraDF)
```

```python
choleraDF.count()
```

```python
choleraDF.isnull().sum()
```

```python
choleraDFclean = choleraDF.dropna().copy()
display(choleraDFclean)
```

```python
choleraDFclean.count()
```

```python
choleraDFclean.isnull().sum()
```

```python
choleraDFclean.describe()
```

```python
choleraDFclean.info()
```

```python
choleraDFclean["Number of reported cases of cholera"] = pd.to_numeric(choleraDFclean["Number of reported cases of cholera"], errors="coerce")
choleraDFclean["Number of reported deaths from cholera"] = pd.to_numeric(choleraDFclean["Number of reported deaths from cholera"], errors="coerce")
choleraDFclean["Cholera case fatality rate"] = pd.to_numeric(choleraDFclean["Cholera case fatality rate"], errors="coerce")

```

```python
choleraDFclean.info()
```

```python
choleraDFclean = choleraDFclean.dropna()
display(choleraDFclean)
```

```python
choleraDFclean.isnull().sum()
```

```python
#Sort dataframes by column names
choleraDFclean.sort_values(by="Number of reported cases of cholera",ascending=False)
```

```python
choleraDFclean.describe()
```

```python
#Scatterplot
sns.regplot(x = "Number of reported cases of cholera", y = "Number of reported deaths from cholera", data = choleraDFclean, line_kws={"color":"green"})
```
